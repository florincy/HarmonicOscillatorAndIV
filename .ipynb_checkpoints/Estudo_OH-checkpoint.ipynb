## Oscilador Harmônico Clássico

Primeiro, vamos estudar o oscilador harmônico clássico. A força restauradora é dada pela Lei de Hooke:

$$ F = -kx $$

A solução da equação do movimento é:

$$ x(t) = A \cos(\omega t + \phi) $$

onde $$\omega = \sqrt{\frac{k}{m}}$$ é a frequência angular.




**Vamos observar a relação entre constante de força e massa, e como tais valores alteram o deslocamento e energia do sistema**

```python
#Definição dos parâmetros
k = float(input("Constante de força (k): "))
m = float(input("Massa (m): "))
```

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Define physical constants
w = np.sqrt(k/m)  # Angular frequency
t = np.linspace(0, 10, 1000)  # Time array

# Define functions
def deslocamento_classico(A, w, t):
    return A * np.cos(w * t)

def energia_potencial(k, x):
    return 0.5 * k * x**2

def energia_cinetica(m, w, A, t):
    return 0.5 * m * (w * A * np.sin(w * t))**2

def energia_total(k, A):
    return 0.5 * k * A**2

# Classical Harmonic Oscillator calculations
x_classico = deslocamento_classico(A, w, t)
V_classico = energia_potencial(k, x_classico)
T_classico = energia_cinetica(m, w, A, t)
E_classico = energia_total(k, A)

# Plot Classical Harmonic Oscillator
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.plot(t, x_classico, label='Deslocamento')
plt.xlabel('Tempo')
plt.ylabel('Deslocamento')
plt.title('Deslocamento do Oscilador Harmônico Clássico')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(x_classico, V_classico, label='Energia Potencial')
plt.xlabel('Deslocamento')
plt.ylabel('Energia Potencial')
plt.title('Energia Potencial do Oscilador Harmônico Clássico')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(t, T_classico, label='Energia Cinética')
plt.xlabel('Tempo')
plt.ylabel('Energia Cinética')
plt.title('Energia Cinética do Oscilador Harmônico Clássico')
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(t, T_classico + V_classico, label='Energia Total')
plt.xlabel('Tempo')
plt.ylabel('Energia Total')
plt.title('Energia Total do Oscilador Harmônico Clássico')
plt.legend()

plt.tight_layout()
plt.show()
```

# Oscilador Harmônico Quântico

## Introdução

O oscilador harmônico quântico é um dos problemas centrais e mais estudados na física quântica e química quântica. Ele serve como um modelo para uma ampla gama de sistemas físicos, desde átomos e moléculas até campos eletromagnéticos. A compreensão detalhada deste modelo proporciona insights profundos sobre o comportamento de sistemas quânticos mais complexos.

## A Equação de Schrödinger para o Oscilador Harmônico

O oscilador harmônico clássico é caracterizado por uma partícula sujeita a uma força restauradora proporcional ao deslocamento da partícula a partir de sua posição de equilíbrio. Em termos quânticos, esse sistema é descrito pela equação de Schrödinger independente do tempo:

$$
-\frac{\hbar^2}{2m} \frac{d^2 \psi}{dx^2} + \frac{1}{2} m \omega^2 x^2 \psi = E \psi
$$

Aqui, $\psi$ é a função de onda da partícula,  m  é a massa da partícula,  $\omega$ é a frequência angular do oscilador, $\hbar$ é a constante de Planck reduzida, $E$ é a energia total do sistema, e $x$ é a posição.

## Soluções da Equação de Schrödinger

As soluções da equação de Schrödinger para o oscilador harmônico são bem conhecidas e consistem em funções de onda $\psi_n(x)$ e níveis de energia $E_n$ que são dados por:

### Funções de Onda

As funções de onda são:

$$
\psi_n(x) = \frac{1}{\sqrt{2^n n!}} \left( \frac{m \omega}{\pi \hbar} \right)^{1/4} e^{-\frac{m \omega x^2}{2 \hbar}} H_n \left( \sqrt{\frac{m \omega}{\hbar}} x \right)
$$

onde $H_n$ são os polinômios de Hermite de ordem $n$. Estes polinômios são soluções da equação diferencial de Hermite, que surge naturalmente ao resolver a equação de Schrödinger para o oscilador harmônico.

### Níveis de Energia

Os níveis de energia são quantizados e dados por:

$$
E_n = \hbar \omega \left( n + \frac{1}{2} \right)
$$

onde \( n \) é um número inteiro não negativo (n = 0, 1, 2, ...). Este resultado mostra que as energias do oscilador harmônico quântico são discretas e igualmente espaçadas.

## Propriedades das Funções de Onda

### Normalização

As funções de onda $\psi_n(x)$ são normalizadas, ou seja, satisfazem a condição:

$$
\int_{-\infty}^{\infty} |\psi_n(x)|^2 \, dx = 1
$$

### Ortogonalidade

As funções de onda de diferentes estados são ortogonais, o que significa que:

$$
\int_{-\infty}^{\infty} \psi_n(x) \psi_m(x) \, dx = 0 \quad \text{para} \quad n \neq m
$$

### Nodos

A função de onda $\psi_n(x)$ possui $n$ nodos (pontos onde $\psi_n(x) = 0$, refletindo a natureza quântica dos estados de energia mais elevados.

## Modelo para Vibrações Moleculares
 O modelo do oscilador harmônico fornece uma aproximação razoável para as vibrações de moléculas diatômicas. As energias vibracionais dos estados de vibração de uma molécula podem ser aproximadas pelos níveis de energia do oscilador harmônico.


 ### Exemplo
 Em uma molécula diatômica, os átomos vibram em torno de sua posição de equilíbrio como um oscilador harmônico. A energia vibracional de uma molécula pode ser quantizada, e os níveis de energia vibracionais podem ser descritos pelos níveis de energia do oscilador harmônico quântico. Assim, as energias vibracionais de uma molécula diatômica são dadas por:

   $$
   E_n = \hbar \omega \left( n + \frac{1}{2} \right)
   $$

  onde \( n \) é o número quântico vibracional. Isso ajuda a explicar os espectros vibracionais observados em espectroscopia infravermelha e Raman. A aplicação das equações do oscilador harmônico para casos como esse é similiar ao introduzido, apenas substituímos a massa pela massa reduzida do sistema.

```python
#GERAR GRÁFICOS PARA MOLÉCULAS DIATÔMICAS
import numpy as np
import matplotlib.pyplot as plt
import itertools

def massa_reduzida(m1, m2):
    return (m1 * m2) / (m1 + m2)

def frequencia_vibracional(k, mu):
    return (1 / (2 * np.pi)) * np.sqrt(k / mu)

def energia_vibracional(n, nu):
    h = 6.62607015e-34  # Constante de Planck em J·s
    return (n + 0.5) * h * nu

def energia_potencial(k, x):
    return 0.5 * k * x**2

def plotar_potencial_vibracional(cor_linhas):
    # Solicitar os parâmetros ao usuário
    nome = input("Nome do átomo: ")
    m1 = float(input("Massa do átomo (kg): "))
    k = float(input("Constante de força (N/m): "))
    m2 = 1.67e-27  # kg (massa aproximada de um átomo de hidrogênio)

    # Calcular a massa reduzida
    mu = massa_reduzida(m1, m2)

    # Frequência vibracional da molécula diatômica
    nu = frequencia_vibracional(k, mu)

    # Plotar Potencial Vibracional
    x_vibracao = np.linspace(-1e-10, 1e-10, 1000)  # Deslocamento em metros
    V_vibracao = energia_potencial(k, x_vibracao)

    plt.plot(x_vibracao, V_vibracao, label=f'Potencial Vibracional ({nome})', color=cor_linhas)

    # Calcular e plotar níveis de energia vibracionais
    niveis_vibracionais = [energia_vibracional(n, nu) for n in range(3)]  # Ajuste o range conforme necessário
    for n, energia in enumerate(niveis_vibracionais):
        plt.hlines(energia, x_vibracao[0], x_vibracao[-1], colors=cor_linhas, linestyles='--', label=f'Nível {n} ({nome})' if n == 0 else "")

#RODE ESSE SCRIPT PARA GERAR GRÁFICOS PARA SUAS MOLÉCULAS DIATÔMICAS
# Configuração inicial do gráfico
plt.figure(figsize=(8, 6))
plt.xlabel('Deslocamento (m)')
plt.ylabel('Energia Potencial (J)')
plt.title('Potencial Vibracional de uma Molécula Diatômica com Níveis de Energia')

# Gerador de cores
cores = itertools.cycle(plt.cm.tab10.colors)  # Usar uma paleta de cores padrão

# Loop para permitir a adição de múltiplas linhas
while True:
    cor_linhas = next(cores)  # Pega a próxima cor
    plotar_potencial_vibracional(cor_linhas)
    plt.legend()
    plt.show()
    
    continuar = input("Deseja adicionar mais uma linha? (s/n): ")
    if continuar.lower() != 's':
        break
```


 ### A Física por trás do infravermelho

 Consideramos que a absorção/emissão de radiação por uma sistema ocorre devido a variação no seu momento de dipolo, de forma que a frequência da radiação é idêntica à oscilação do dipolo. O momento de dipolo depende da configuração nuclear e é muda de acordo com a vibração do sistema. No caso das moléculas diatômicas, a única coordenada normal do sistema coincide com a coordenada interna da ligação, representada por $q$. Expandindo o momento de dipolo em série de Taylor em função da coordenada $q$:
 $$
 \mu = \mu_0 + \left( \frac{d\mu}{dx} \right)_0 + \ldots
 $$
 Truncamos a série no segundo termo e, assim, observamos a **condição que determina absorção no infravermelho** (Tal equação deve ser válida para pelo menos uma das componentes do momento de dipolo)
 $$
 \left( \frac{d\mu}{dx} \right)_0 \neq 0
 $$
 No tocante a intensidade do sinal em infravermelho, podemos associa-lá a soma dos quadrados dos componentes do momento de transição de dipolo $({|\mu_{mn}|}^2)$, que é calculado a partir de dois estados $\psi_m$ e $\psi_n$ para cada componente $x,y,z$ dessas respectivas funções.
 $$
 \mu_{mn} = \int \psi_m \mu \psi_n \, d\tau
 $$
 Essa medida corresponde fisicamente ao movimento dos elétrons durante a transição entre os estados $\psi_m$ e $\psi_n$. Tais definições permitem estabelecer que **transições permitdas devem apresentar, para pelo menos umas componentes**, essa é uma das chamadas **regras de seleção**:
 $$
 \mu_{mn} \neq 0
 $$
 Substituindo a expressão para $\mu$ em $\mu_mn$:
 $$
 \mu_{mn} = \mu_0 \int \psi_m \psi_n \, d\tau + \left( \frac{d\mu}{dx} \right)_0 \cdot \int \psi_m  q \psi_n \, d\tau
 $$
 O primeiro termo é 0, devido propriedadade da ortogonalidade das funções de onda, assim, para a expressão não ser nula, o segundo termo deve apresentar:
 1.  $\left( \frac{d\mu}{dx} \right)_0 \neq 0$, conforme já mencionada acima
 2.  $\int \psi_m  q \psi_n \, d\tau \neq 0$, de forma que o produto $\psi_m q \psi_n$ deve ser função par. Sabendo que q é ímpar, $\psi_m \psi_n$ deve ser ímpar, ou seja, as funções de onda devem ser de paridade distinta. A partir dessa condição, é definida a regra de seleção para o espectro IV:
 $$
 \Delta v \pm 1
 $$
 


### Atividade da vibrações e os modos normais
 As vibrações que encontramos nos sistemas moleculares estão diretamente associadas translações, o que permite a aplicação da teoria de grupos para interpretação dessas vibrações. Nesse sentido, temos que os 3N-6 graus internos de liberdade de uma molécula apresentam 3N-6 correspondentes **modos vibracionais normais independentes**. Para cada modo vibracional, a frequência em que os átomos oscilam é constante e todos os átomos passam simultaneamente por suas posições de equilíbrio. Apesar das amplitudes vibracionais relativas dos átomos individualmente serem diferentes, o centro de gravidade do sistema não se move, assim, a molécula não realiza rotação. Numa vibração harmônica, as forças que mantêm a molécula estruturalmente coesa são funções lineares do deslocamento dos núcleos e, portanto, das configurações de equilíbrio dos átomos. Compreendendo melhor a relação entre vibrações e movimentos internos do sistema, associamos a atividade no infravermelho como a ressonância entre a frequência da radiação "pertubadora" e algum/s modo/s vibracionais normais da molécula.

 <img src="Infravermelho_Modos.png" alt="Modelo" width="200"/>
 
 Os componentes do vetor de momento dipolar x,y,z possuem as propriedades de simetria dos vetores translacionais $T_x , T_y , T_z $, logo, possuem os mesmos caracteres. Dessa maneira, qualquer espécie de simetria que possa ser ativa no IV pertence à terceira columa da tabela de caracteres e transformam os vetores translacionais.

 Tabela: Tabela de caracteres para o grupo pontual $C_{2v}$

   | $C_{2v}$   | E | $C_2$  | $\sigma_x$   |  $\sigma_{yz}$  | Graus de liberdade |
   |------------|---|--------|--------------|-----------------|--------------------|
   | $A_1$      | 1 | 1      | 1            | 1               | $T_z, V_1, V_2$    |
   | $A_2$      | 1 | 1      | -1           | -1              | $R_z$              |
   | $B_1$      | 1 | -1     | 1            | -1              | $T_x, R_y$         |
   | $B_2$      | 1 | -1     | -1           | 1               | $T_y, R_x, V_3$    |
   

   Portanto, dado um espectro de infravermelho, podemos atribuir as frequências absorvidas (ou "números de onda") aos modos normais vibracionais existes no sistema químico. Para isso, a química computacional pode ser empregada.

```python
#RODA PARA GERAR O ESPECTRO IV TEÓRICO
'''
# orca-ir
'''

import sys                              # sys files processing
import os                               # os file processing
import re                               # regular expressions
import numpy as np                      # summation
import matplotlib.pyplot as plt         # plots
from scipy.signal import find_peaks     # peak detection

# global constants
found_ir_section = False                # check for IR data in out
specstring_start = 'IR SPECTRUM'        # check orca.out from here
specstring_end = 'The first'            # stop reading orca.out from here
w = 15                                  # w = line width for broadening, FWHM
wn_add = 150                            # add +150 to spectra x (required for convolution)

# plot config section - configure here
high_to_low_wn = True                   # go from high to low wave number, normal for IR spectra, low wn to high wn if False
transm_style = True                     # show spectra in transmittance style, absorption style if False
show_grid = True                        # show grid if True
show_conv_spectrum = True               # show the convoluted spectra if True (if False peak labels will not be shown)
show_sticks = True                      # show the stick spectra if True
show_single_gauss = False               # show single gauss functions if True
show_single_gauss_area = False          # show single gauss functions - area plot if True
label_rel_pos_y = -15                   # -15 for transmittance style, 5 for absorption style 
save_spectrum = True                    # save spectrum if True
show_spectrum = False                   # show the matplotlib window if True
label_peaks = True                      # show peak labels if True
minor_ticks = True                      # show minor ticks if True
linear_locator = True                   # tick locations at the beginning and end of the spectrum x-axis, evenly spaced
spectrum_title = "IR spectrum"          # title
spectrum_title_weight = "bold"          # weight of the title font: 'normal' | 'bold' | 'heavy' | 'light' | 'ultrabold' | 'ultralight'
y_label_trans = "transmittance"         # label of the y axis - Transmittance
y_label_abs = "intensity"               # label of y-axis - Absorption
x_label = r'$\tilde{\nu}$ /cm$^{-1}$'   # label of the x-axis
figure_dpi = 300                        # DPI of the picture

# global lists
modelist = []         # mode
freqlist = []         # frequency
intenslist = []       # intensity absolute
gauss_sum = []        # list for the sum of single gaussian spectra = the convoluted spectrum

def roundup(x):
    # round to next 100
    return x if x % 100 == 0 else x + 100 - x % 100

def gauss(a, m, x, w):
    # calculation of the Gaussian line shape
    # a = amplitude (max y, intensity)
    # x = position
    # m = maximum/meadian (stick position in x, wave number)
    # w = line width, FWHM
    return a * np.exp(-(np.log(2) * ((m - x) / w) ** 2))

# Define the ORCA output file path directly
filename = '/home/nqtcm11/Documentos/Simone/Self-Study-For-QM/output.out'  

# open a file
# check existence
try:
    with open(filename, "r") as input_file:
        for line in input_file:
            # start extract text 
            if "Program Version 6" in line:
                # thanks to the orca prgrmrs intensity is now in a different column
                intens_column = 3
            if "Program Version 4" in line:
                intens_column = 2
            if "Program Version 3" in line:
                intens_column = 2
            if line.startswith(specstring_start):
                # found IR data in orca.out
                found_ir_section = True
                for line in input_file:
                    # stop extract text 
                    if line.startswith(specstring_end):
                        break
                    # only recognize lines that start with number
                    # split line into 3 lists mode, frequencies, intensities
                    # line should start with a number
                    if re.search("\d:", line): 
                        modelist.append(int(line.strip().split(":")[0])) 
                        freqlist.append(float(line.strip().split()[1]))
                        intenslist.append(float(line.strip().split()[intens_column]))

# file not found -> exit here
except IOError:
    print(f"'{filename}'" + " not found")
    sys.exit(1)

# no IR data in orca.out -> exit here
if not found_ir_section:
    print(f"'{specstring_start}'" + "not found in" + f"'{filename}'")
    sys.exit(1)

# prepare plot
fig, ax = plt.subplots()

# plot range in x
plt_range_x = np.arange(0, max(freqlist) + wn_add, 1)

# plot single gauss function for every frequency freq
# generate summation of single gauss functions
for index, freq in enumerate(freqlist):
    # single gauss function line plot
    if show_single_gauss:
        ax.plot(plt_range_x, gauss(intenslist[index], plt_range_x, freq, w), color="grey", alpha=0.5) 
    # single gauss function filled plot
    if show_single_gauss_area:
        ax.fill_between(plt_range_x, gauss(intenslist[index], plt_range_x, freq, w), color="grey", alpha=0.5)
    # sum of gauss functions
    gauss_sum.append(gauss(intenslist[index], plt_range_x, freq, w))

# y values of the gauss summation
plt_range_gauss_sum_y = np.sum(gauss_sum, axis=0)

# find peaks scipy function, change height for level of detection
peaks, _ = find_peaks(plt_range_gauss_sum_y, height=0.1)

# plot spectra
if show_conv_spectrum:
    ax.plot(plt_range_x, plt_range_gauss_sum_y, color="black", linewidth=0.8)

# plot sticks
if show_sticks:
    ax.stem(freqlist, intenslist, "dimgrey", markerfmt=" ", basefmt=" ")

# optional mark peaks - uncomment in case
# ax.plot(peaks, plt_range_gauss_sum_y[peaks], "|")

# label peaks
# show peak labels only if the convoluted spectrum is shown (first if)
if show_conv_spectrum:  
    if label_peaks:
        if not transm_style:
            label_rel_pos_y = 5   # in case of absorption style spectra
            
        for index, txt in enumerate(peaks):
            
            if transm_style:
                # corr_factor - maintain distance from label to peak in transmittance style
                # sensitive to peak label font size
                corr_factor = (4 - len(str(peaks[index]))) * 3.75
                # if one does not care:
                # corr_factor = 0 
            else:
                # not necessary for absorption style
                corr_factor = 0
                
            ax.annotate(peaks[index], xy=(peaks[index], plt_range_gauss_sum_y[peaks[index]]), ha="center", rotation=90, size=6,
                        xytext=(0, label_rel_pos_y + corr_factor), textcoords='offset points')
    
ax.set_xlabel(x_label)                                          # label x axis

if transm_style:
    ax.set_ylabel(y_label_trans)                                # label y axis
else:
    ax.set_ylabel(y_label_abs)
    
ax.set_title(spectrum_title, fontweight=spectrum_title_weight)  # title
ax.get_yaxis().set_ticks([])                                    # remove ticks from y axis
plt.tight_layout()                                              # tight layout
if minor_ticks:
    ax.minorticks_on()                                          # show minor ticks

# x, y axis manipulation for transmittance or absorption style 
if transm_style:
    plt.ylim(max(plt_range_gauss_sum_y) + max(plt_range_gauss_sum_y) * 0.1, 0) # +10% for labels
else:
    plt.ylim(0, max(plt_range_gauss_sum_y) + max(plt_range_gauss_sum_y) * 0.1) # +10% for labels
    
if high_to_low_wn:
    plt.xlim(roundup(max(plt_range_x)), 0) # round to next 100
else:
    plt.xlim(0, roundup(max(plt_range_x)))

# tick locations at the beginning and end of the spectrum x-axis, evenly spaced
if linear_locator:
    ax.xaxis.set_major_locator(plt.LinearLocator())

# show grid
if show_grid:
    ax.grid(True, which='major', axis='x', color='black', linestyle='dotted', linewidth=0.5)

# increase figure size N times
N = 1.5
params = plt.gcf()
plSize = params.get_size_inches()
params.set_size_inches((plSize[0] * N, plSize[1] * N))

# save the plot
if save_spectrum:
    filename, file_extension = os.path.splitext(filename)
    plt.savefig(f"{filename}-ir.png", dpi=figure_dpi)

# show the plot
if show_spectrum:
    plt.show()
    
```

<b>NOTA:</b>

```python
import os
from ase.io import read, write
from ase.visualize import view
import re

# Directory containing the .xyz files
xyz_directory = '/home/nqtcm11/Documentos/Simone/Self-Study-For-QM/'  # Update with your directory path

def extract_frequencies(directory):
    frequencies = {}
    for file_name in os.listdir(directory):
        if file_name.endswith('.xyz'):
            file_path = os.path.join(directory, file_name)
            try:
                with open(file_path, 'r') as f:
                    frequency_found = False
                    for line in f:
                        if "Frequency" in line and not frequency_found:
                            match = re.search(r'Frequency\(\d+\) (\d+\.\d+)', line)
                            if match:
                                frequency = float(match.group(1))
                                frequencies[file_name] = f"{frequency:.2f}"
                                frequency_found = True
                                break
            except Exception as e:
                print(f"Error reading file {file_name}: {e}")
    return frequencies

# Extract frequencies from all .xyz files
frequencies = extract_frequencies(xyz_directory)

# Allow user to select a file by frequency
selected_frequency = input("Insira a frequência do modo a ser visualizado: ")

# Find the file corresponding to the selected frequency
selected_file = None
for file_name, frequency in frequencies.items():
    if frequency == selected_frequency:
        selected_file = file_name
        break

if selected_file:
    # Proceed with further processing or visualization of the selected file
    # Define file paths
    input_xyz = os.path.join(xyz_directory, selected_file)
    output_traj = os.path.join(xyz_directory, f"{os.path.splitext(selected_file)[0]}.traj")  # Remove .xyz and add .traj

    # Read all frames from the XYZ file
    try:
        frames = read(input_xyz, index=':')  # Read all frames
    except Exception as e:
        print(f"Error reading the file: {e}")
        frames = None

    if frames:
        # Write frames to a .traj file
        try:
            write(output_traj, frames)  # Write frames to .traj file
        except Exception as e:
            print(f"Error writing the .traj file: {e}")

        # Ensure the .traj file exists before trying to read it
        if os.path.exists(output_traj):
            # Read the .traj file and visualize
            try:
                frames = read(output_traj, index=':')
                # Read all frames from the .traj file
                view(frames)  # Visualize the frames
            except Exception as e:
                print(f"Error reading or visualizing the .traj file: {e}")
        else:
            print(f"The .traj file does not exist: {output_traj}")
else:
    print("No file found with the selected frequency.")
```
